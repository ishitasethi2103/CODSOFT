```python
movies['profit']=movies['revenue']-movies['budget']
plt.hist(movies.profit,histtype='bar')
plt.xlabel("profit in  10crores")
plt.ylabel("no of movies")
```

```python
moviecredits_column_renamed=moviecredits.rename(index=str,columns={"movie_id":"id"})
movies_merge=movies.merge(moviecredits_column_renamed, on='id')
movies_merge.head()
```

```python
movies_clean=movies_merge.drop(columns=['homepage','title_x','title_y','status','production_countries'])
movies_clean.head()
```

```python
movies_clean.info()
```

```python
v=movies_clean['vote_count']
R=movies_clean['vote_average']
C=movies_clean['vote_average'].mean()
m=movies_clean['vote_average'].quantile(0.80)
movies_clean['weighted average']=((R*v)+(C*m))/(v+m)
movies_clean.head()
```

```python
movie_sorted_ranking=movies_clean.sort_values('weighted average',ascending=False)
movie_sorted_ranking[['original_title','vote_count','vote_average','weighted average','popularity']].head(10)
```

```python
weight_average=movie_sorted_ranking.sort_values('weighted average',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_average['weighted average'].head(10),y=weight_average['original_title'].head(10))
plt.xlim(4,10)
plt.title('Best movies acc to avg votes:',weight='bold')
plt.xlabel('Average weighted score:',weight='bold')
plt.ylabel('Movie title:',weight='bold')
plt.savefig('best_movies.png')
```

```python
popularity=movie_sorted_ranking.sort_values('popularity',ascending=False)
plt.figure(figsize=(12,6))
ax=sns.barplot(x=popularity['popularity'].head(10),y=popularity['original_title'].head(10),data=popularity['popularity'].head(10))
plt.title('Most popular by votes:',weight='bold')
plt.xlabel('Score of popularity:',weight='bold')
plt.ylabel('Movie title:',weight='bold')
plt.savefig('best_popular_movies.png')
```

```python
from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler()
movie_scale = scaling.fit_transform(movies_clean[['weighted average', 'popularity']])
movie_normalized_df = pd.DataFrame(movie_scale, columns=['weighted average', 'popularity'])
movie_normalized_df.head()
```

```python
movies_clean[['normalized_weight_average','normalized_popularity']]=movie_normalized_df
movies_clean.head()
```

```python
movies_clean['score']=movies_clean['normalized_weight_average']*0.4+movies_clean['normalized_popularity']*0.6
movies_score=movies_clean.sort_values(['score'],ascending=False)
movies_score[['original_title','normalized_weight_average','normalized_popularity','score']].head(20)
```

```python
scored=movies_clean.sort_values('score',ascending=False)
plt.figure(figsize=(16,6))
ax=sns.barplot(x=scored['score'].head(10),y=scored['original_title'].head(10),data=scored['popularity'].head(10))
plt.title('Best rated and most popular',weight='bold')
plt.xlabel('Score',weight='bold')
plt.savefig('scored_movies.png')
```
